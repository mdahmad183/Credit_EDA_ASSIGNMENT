## * Funtion to pllot the univariate categorical variables

```python
def plotuniNA_(var):
    
    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(18,6))
    
    sns.countplot(x=var, data=NA_0,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {var} for Non-Defaulters', fontsize=12)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
    
# Adding the normalized percentage 
    
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(NA_0))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=var, data= NA_1 , ax=ax2)
    ax2.set_ylabel('Total couns')
    ax2.set_title(f'Distribution of {var} for Defaulters' , fontsize=12)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
    
# Adding the normalized percentage
    
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(NA_1))*100),(p.get_x()+0.1, p.get_height()+50))
        
    plt.show()
```

# 2.4.5 - Bi-Variate Analysis of numarical variables

## * use Scatter plot for Continuous variables

```python
def plotbivar(var1,var2):
    
    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.scatterplot(x=var1, y=var2,data=NA_0,ax=ax1)
    ax1.set_xlabel(var1)
    ax1.set_ylabel(var2)
    ax1.set_title(f'{var1} vs {var2} for Non-Defaulters', fontsize=15)
    
    sns.scatterplot(x=var1, y=var2, data=NA_1,ax=ax2)
    ax2.set_xlabel(var1)
    ax2.set_ylabel(var2)
    ax2.set_title(f'{var1} vs {var2} for Defaulters', fontsize=15)
    
    plt.show()
```

```python
plotbivar('AMT_CREDIT','CNT_FAM_MEMBERS')
```

* we can see that the density in the lower left corner is similar in both the case ,

# 3.5 - Using Box-plot for Bi-variate analysis on Categorical Vs. Numeric Columns

   * by Variant analysis function

```python
def plot_by_cat_num(cat, num):
    
    plt.style.use('ggplot')
    sns.despine
    fig,ax = plt.subplots(1,1,figsize=(10,8))
    
    sns.boxplot(x=cat,y=num, data=PA)
    ax.set_xlabel(f'{num}')
    ax.set_xlabel(f'{cat}')
    
    ax.set_title(f'{cat} Vs {num}', fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 45 , ha= "right")
    
    plt.show()
```

```python
# by varient analysis of NAME_CONTRACT_STATUS & AMT_ANNUITY from PA .
```

```python
plot_by_cat_num("NAME_CONTRACT_STATUS" , "AMT_ANNUITY")
```

* the loan Applicaton for people with lower AMT_ANNUITY get canceled or Unused most of the time.

* also the Application with too high AMT_ANNUITY also got refused more often then other.

```python
# by varient analysis of NAME_CONTRACT_STATUS & AMT_CREDIT from PA .
```

```python
plot_by_cat_num("NAME_CONTRACT_STATUS" , "AMT_CREDIT")
```

* We can infer that when the AMT_CREDIT is too low , it get's cancelled / unused most of the time.

```python
def plotuni_combined(Varx , Vary):              # 100% BAR CHART
    plt.style.use('ggplot')
    sns.despine
    New_Data = Final_Application.pivot_table(values='SK_ID_CURR',
                                           index=Varx,
                                           columns=Vary,
                                           aggfunc='count')
    New_Data = New_Data.div(New_Data.sum(axis=1),axis='rows')*100
    sns.set()
    New_Data.plot(kind='bar' , stacked=True,figsize=(15,5))
    plt.title(f'Effect of {Varx} on Loan Approval')
    plt.xlabel(f'{Varx}')
    plt.ylabel(f'{Vary}%')
    
    plt.show()
```

```python
plotuni_combined('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')
```
